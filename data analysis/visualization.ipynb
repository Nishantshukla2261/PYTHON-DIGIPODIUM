```python
t=np.linspace(0,3*np.pi,50)
x=np.sin(t)
y=np.cos(t)
```

```python
plt.figure(figsize=(15,6))
plt.plot(t,x,'ro--') ## o and x are predefined o- bhi likh sakte h
plt.plot(t,y,'bx--',markersize=5) ## zoom karega ye
```

```python
plt.figure(figsize=(15,7))
plt.plot(t,x,'ko:') ## o and x are predefined o- bhi likh sakte h ## k means black
plt.plot(t,y,'h',markersize=5) ## zoom karega ye
```

```python
plt.figure(figsize=(15,7))
plt.plot(t,x,marker='*',color='#903399',markersize=15,markeredgecolor='blue',
         markerfacecolor='yellow')  ## (red green blue))o and x are predefined o- bhi likh sakte h ## k means black
plt.plot(t,y,'h',markersize=10) ## zoom karega ye
```

```python
plt.figure(figsize=(10,6)) ## sixe ghata badha sakte h
plt.plot(years,prices,'ro--') ## ro means red -- means line me lana
plt.title('Prices of stocks') ## upr so kr sakte h
plt.xlabel("Years") ## corner pe likh k aayega
plt.ylabel("Prices")
plt.hlines(140,2010,2015)  
plt.hlines([140,140,150],## 140 se sure 2010 se start kr k 2012 tk jayega verticaly value le rha h
           [2010,2012,2010],
           [2012,2015,2013],
            linestyles=['dashed','solid'])
plt.vlines([2012,2013],
           [min(prices),min(prices)],
           [140,150],
           linestyles='dotted',
           colors=['red','blue'],
           label='price')
```
